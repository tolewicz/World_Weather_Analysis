```python
import pandas as pd
import numpy as np
from citipy import citipy
```

```python
#getting coordinates by using random numbers between -90,90 and -180,180

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs) 
```

```python
# Add the latitudes and longitudes to a list.
#we have a list named coordinates where random coordinates are put

coordinates = list(lat_lngs)

#checking how many coordinates we get - 1500
len(coordinates)
```

```python
# Use the citipy module to determine city based on latitude and longitude.
#the module check if there is any city nearby 

from citipy import citipy
```

```python
# Use the tuple() function to display the latitude and longitude combinations.
#if there is no city nearby the coordinate is dropped

for coordinate in coordinates:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
# Create a list for holding the cities. so if there is no city it will not show in the list 
cities = []

# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
        
# Print the city count to confirm sufficient count.
#so the number of entries was decreased from 1500 to ~500
len(cities)
```

```python
# Import the requests library.
import requests

# Import the API key.
#I got a separate python program named config. this command request accessing the program and getting 
#"weather_api_key" variable which has my key

from config import weather_api_key
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
# Import the datetime module from the datetime library.
from datetime import datetime
```

```python
# Create an empty list to hold the weather data.

city_data = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1
```

```python
# Import the time module.
import time
# Get today's date in seconds.
today = time.strftime("%x")
today
```

```python
# Loop through all the cities in the list.
#we gave two iterators i (for record number) and city,we iterate on cities list
#we use "enumerate" because we cycle over the dictionary   

city_found = 0
city_not_found = 0

for i, city in enumerate(cities):


    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1 #every 50 the set count is +1
        record_count = 1 #every 50 record count is reset to 1
        
    # set endpoint URL with each city.
    city_url = url + "&q=" + city

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add 1 to the record count.
    record_count += 1
    

    # Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_weather_dexcription = city_weather["weather"][0]["description"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
  
        try:
            rain = city_weather["rain"]["3h"]
        except:
            rain = 0
        try:
            snow = city_weather["snow"]["3h"]
        except:
            snow = 0
        
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Weather description":city_weather_dexcription,
                          "Rain": rain,
                          "Snow": snow,
                          "Country": city_country,
                          "Date": city_date})
        city_found +=1

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        city_not_found +=1
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
print(len(city_data))
print(f'city found {city_found}')
print(f'skipped record {city_not_found}')
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df
```

```python
#City, Country, Date, Lat, Lng, Max Temp, Humidity, Cloudiness, and Wind Speed
new_order =["City", "Country", "Date", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Weather description", "Rain","Snow"]
city_data_df = city_data_df[new_order]
city_data_df
```

```python
# Create the output file (CSV) with city list
import os

output_data_file = os.path.join("Data","WeatherPy_challenge.csv")

# Export the City_Data into a CSV.
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
